```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NCKH/DATA.csv')
```

```python
df = pd.DataFrame(data)
```

```python
for column in df.columns:
    unique_values = df[column].unique()
    count_unique = len(unique_values)
    print(f"Column: {column}, Unique Values: {unique_values}, Count: {count_unique}")
```

```python
# Tạo DataFrame mới để chứa thông tin không lặp lại
unique_info = pd.DataFrame(columns=['Column', 'Unique_Values'])

# Đếm và lưu thông tin không lặp lại của mỗi cột vào DataFrame mới
for column in df.columns:
    unique_values = df[column].unique()
    unique_info = pd.concat([unique_info, pd.DataFrame({column: unique_values})], axis=1)

# Xuất DataFrame chứa thông tin không lặp lại ra file CSV
unique_info.to_csv('/content/drive/MyDrive/Colab Notebooks/NCKH/unique_info.csv', index=False)
```
