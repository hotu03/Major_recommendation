## A Levels (Advanced Level):
### Là một hệ thống giáo dục tiên tiến được áp dụng tại nhiều quốc gia, bao gồm cả Pakistan và Ấn Độ. Học sinh học chương trình A Levels sau khi hoàn thành giáo trình trung học cơ sở. Đây là một bậc học được coi là cao cấp và chuẩn bị cho việc học ở cấp đại học.

##Fsc Pre Engineering (Faculty of Science, Pre-Engineering):
### Một hệ thống giáo dục chia thành các khoa (faculties), trong trường hợp này là khoa Khoa học, Trước kỹ thuật (Pre-Engineering). Học sinh theo học chương trình này chuẩn bị cho các ngành kỹ thuật tại cấp đại học.

##I COM (Intermediate in Commerce):
### Một chương trình học tập ở mức trung cấp liên quan đến lĩnh vực Thương mại. Học sinh I COM học các môn như Kế toán, Tài chính và Quản trị kinh doanh, chuẩn bị cho sự nghiệp trong lĩnh vực kinh doanh và tài chính.

##Fsc Pre Medical (Faculty of Science, Pre-Medical):
### Tương tự như Fsc Pre Engineering, nhưng hướng đến các ngành y khoa. Học sinh theo học chương trình này thường chuẩn bị cho việc theo đuổi các chuyên ngành y khoa như Y học, Nha khoa, hoặc Dược học.

##ICS (Intermediate in Computer Science):
### Chương trình học về Khoa học máy tính ở mức độ trung cấp. Học sinh ICS học các môn như Lập trình, Cơ sở dữ liệu, và Hệ thống thông tin, để chuẩn bị cho sự nghiệp trong lĩnh vực Công nghệ thông tin.

# Áp dụng mô hình học máy

```python

from google.colab import drive
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import joblib
import numpy as np

drive.mount('/content/drive')

train=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NCKH/Data/Train_data.csv')

# Load mô hình từ file
loaded_svm_model = joblib.load('/content/drive/MyDrive/Colab Notebooks/NCKH/Model/SVM_Model.pkl')
loaded_nb_model = joblib.load('/content/drive/MyDrive/Colab Notebooks/NCKH/Model/NB_Model.pkl')
loaded_knn_model = joblib.load('/content/drive/MyDrive/Colab Notebooks/NCKH/Model/KNN_Model.pkl')


# Tạo dataframe cho dữ liệu dự đoán
new_data = pd.DataFrame({
    'Major Subjects in Matric': [KhoiChuyen],
    'Major in Intermediate': [HeThong],
    'Intermediate Percentage': [ThanhTich],
    'Favorite Subject': [MonHoc],
    'Area of Interest': [LinhVuc]
})

label_encoder_models = {}
for col in train.columns:
    if train[col].dtype == 'object':
        label_encoder = LabelEncoder()
        train[col] = label_encoder.fit_transform(train[col])
        label_encoder_models[col] = label_encoder
        joblib.dump(label_encoder, f'/content/drive/MyDrive/Colab Notebooks/NCKH/Model/Model_Encoded/label_encoder_{col}.pkl')
```

```python
for col in new_data.columns:
    if col in label_encoder_models:
        label_encoder = label_encoder_models[col]
        new_data[col] = label_encoder.transform(new_data[col])
print(new_data)
```

```python
def decode_label(input_number):
    # Đọc dữ liệu từ file CSV
    df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NCKH/Data/encoded_labels.csv')

    # Tìm nhãn tương ứng với số hiệu đầu vào
    decoded_label = df[df['Số hiệu'] == input_number]['Nhãn'].values

    # Kiểm tra xem số hiệu có tồn tại trong dữ liệu không
    if len(decoded_label) > 0:
        return decoded_label[0]
    else:
        return "Không tìm thấy số hiệu"
```

```python
def result(model):
  import warnings
  warnings.filterwarnings("ignore")
  prediction = loaded_svm_model.predict(new_data)[0]
  print("Predicted Result:", decode_label(prediction))
```
